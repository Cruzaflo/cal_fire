```python
#dependencies
import pandas as pd
import plotly as py
import plotly.express as px
```

```python
#read in data
fire_map_data = pd.read_csv("mapdataall.csv")
fire_map_data.head()
```

```python
#fire IDs per county
fire_map_data.groupby(by="incident_county")["incident_id"].count().to_frame().reset_index()
fire_count_by_county = fire_map_data.groupby(by="incident_county")["incident_id"].count().to_frame().reset_index()

```

```python
fires_by_county_fig = px.bar(fire_count_by_county, x='incident_county', y='incident_id', labels={
                     "incident_county": "County",
                     "incident_id": "Fire Count",
                 }, title = "Fires by County")
               
fires_by_county_fig.update_xaxes(tickangle = 45)

fires_by_county_fig

fires_by_county_fig.write_html("./fires_by_county_fig.html")
```
